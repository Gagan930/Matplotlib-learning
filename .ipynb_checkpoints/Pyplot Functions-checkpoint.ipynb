```python
import matplotlib.pyplot as plt
```

```python
months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]

plt.plot(months,sales,color = 'green',linestyle = '--',marker = 'o',linewidth = 2,label = "2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.title("Monthly Sales Data Report")

# to add graphical grid in the background of your report 
# plt.grid()

plt.grid(color = 'red',linestyle = ':',linewidth = 2)


## to specify the ranges of x-axis and y-axis
plt.xlim(1,4)        # 1 to 4
plt.ylim(0,2000)     # 0 to 2000

plt.legend()
plt.show()
```

### Axis ticks : xticks and yticks

xticks() and yticks() are functions used to get or set the locations and labels of the tick marks on the x-axis and y-axis.

```python
months = [1, 2, 3, 4]
sales = [1000, 1500, 1200, 1800]

plt.plot(months,sales)

# now we will set labels to the numerical values of the X axis using xticks
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])


plt.show()
```

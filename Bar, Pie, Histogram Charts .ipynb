## Bar graph

```python
import matplotlib.pyplot as plt
```

```python
product = ['A','B','C','D']
sales = [1000, 1500, 800, 2000]

plt.bar(product,sales,color = 'hotpink',width = 0.5, label = "2025 sales")
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()


plt.show()
```

```python
# To make horizontal bar chart horizontal use 'barh'

plt.barh(product,sales,color = 'hotpink', label = "2025 sales")
```

## Pie Chart

```python
import matplotlib.pyplot as plt

# 1. Define the data and labels
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [35, 25, 25, 15]  # Proportional sizes for each slice
custom_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# 2. Highlight a specific slice (e.g., 'Apples' offset by 0.1)
explode = (0.1, 0, 0, 0)  

# 3. Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    sizes, 
    explode=explode, 
    labels=categories, 
    colors=custom_colors,
    autopct='%1.1f%%',       # Format percentages to 1 decimal place
    shadow=True,             # Add a subtle drop shadow
    startangle=90            # Rotate start angle to 90 degrees (top center)
)

# 4. Final touches
plt.title('My Favorite Fruits')
plt.axis('equal')            # Ensures the pie chart is a perfect circle
plt.show()
```

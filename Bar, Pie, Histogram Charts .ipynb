## Bar graph

```python
import matplotlib.pyplot as plt
```

```python
product = ['A','B','C','D']
sales = [1000, 1500, 800, 2000]

plt.bar(product,sales,color = 'hotpink',width = 0.5, label = "2025 sales")
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()


plt.show()
```

```python
# To make horizontal bar chart horizontal use 'barh'

plt.barh(product,sales,color = 'hotpink', label = "2025 sales")
```

```python
regions = ["North","South","East","West"]
values = [3000,2000,1500,1000]

plt.pie(values,labels = regions, autopct = "%1.1f%%",colors = ["gold","hotpink","beige","violet"])
plt.title("Revenue Per Region")
plt.show()
```

## Histogram

This function automatically splits your numerical dataset into intervals (called bins), counts how many data points fall into each interval, and draws the resulting bars

```python
# plt.hist(data, bins-number_of_bins, color='colorname', edgecolor='black')

# intervals m data ko baant deta hai , show krta hai ki kis interval m kitna data hai 

# for example : - 100 students scored some marks , how many fall in the 10-20 marks interval , 20-30 marks interval and so on

# these intervals are called bins and we can specify how many bins (intervals) we want.
```

```python
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(scores, bins=5, color = "hotpink", edgecolor = "green")
plt.xlabel("Score Range")
plt.ylabel("Number of students")
plt.title("Score Distribution of Students")
plt.show()



# height of bar = number of students in that range
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate some dummy data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Optional third and fourth variables for size and color mapping
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50) 

# 2. Create the plot
plt.figure(figsize=(8, 6))

# Execute the scatter command
scatter = plt.scatter(
    x, y, 
    s=sizes,          # Varying marker sizes
    c=colors,         # Varying marker colors
    cmap='viridis',   # Smooth color transition theme
    alpha=0.6,        # Translucent dots to see overlaps
    edgecolors='w',   # White borders around dots
    linewidths=1      # Thickness of the border
)

# 3. Add details and polish
plt.title('Customized Matplotlib Scatter Plot', fontsize=14)
plt.xlabel('X Coordinate Axis', fontsize=12)
plt.ylabel('Y Coordinate Axis', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Add a color bar to show what the colors mean
plt.colorbar(scatter, label='Intensity Scale')

# 4. Display the chart
plt.show()
```

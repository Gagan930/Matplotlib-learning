```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

# creating line plot 
plt.plot(x,y,color = "hotpink",marker = "o")
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# saving the plot
plt.savefig("line_plot.png",dpi=300,bbox_inches = "tight")

plt.show()
```

## Subplots

When we want to show multiple graphs in one single canvas, we make subplots

Syntax : 

plt.subplot(nrows, ncols, index) 

plt.plot(x,y)

```python
# plotting a line graph, a scatter plot, histogram and a bar plot in the same canvas
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.subplot(2,2,1)    # 1st position
plt.plot(x,y)
plt.title("Line Chart")


plt.subplot(2,2,2)    # 2nd position 
plt.scatter(x,y)
plt.title("Scatter Chart")


data = [12, 14, 14, 15, 16, 17, 18, 18, 19, 21]

plt.subplot(2,2,3)   # 3rd position
plt.hist(data,bins = 5, color="hotpink", edgecolor = "green")
plt.title("Hist Chart")


plt.subplot(2,2,4)   # 4th position
plt.bar(x,y, color="hotpink", edgecolor = "green")
plt.title("Bar Chart")


plt.show()
```

### Object Oriented Interface Way : More professional syntax to make subplots

Syntax : fig, ax = plt.subplots(nrows ,ncols, figsize = (width,height))

```python
fig , ax = plt.subplots(1,2, figsize = (10,5))         # fig variable will catch the size, ax will have the dimensional access 

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title("Line Plot")

ax[1].bar(x,y,color = "hotpink",edgecolor = "blue")
ax[1].set_title("Bar Chart")

plt.suptitle("Comparison of Line and Bar Charts")

plt.tight_layout()    # this will make the subplots ennlarged enough to take the whole canvas space, also fixes overlap
plt.show()



# this kind of syntax gives us more control over the subplots, we can access the each subplot individually
```

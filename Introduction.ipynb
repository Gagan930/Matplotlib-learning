```python
import matplotlib.pyplot as plt

x = ['Mon','Tue','Wed','Thu','Fri']
y = [10,15,7,20,12]

plt.plot(x,y)

plt.xlabel('Day of the week')
plt.ylabel('Sales Per Day')

plt.title('Bakery Sales This Week')

plt.show()
```
